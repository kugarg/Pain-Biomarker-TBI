# STEP 3 = Handle missing data (Median imputation)

```python
# List of columns with missing values
column_list = ['Age', 'Pain rating', 'CD3%', 'CD3Total', 'CD8%', 'CD8-Suppr',
               'CD4%', 'CD4-Helper', 'H/S-ratio', 'HgB', 'Platelets', 'Neutrophils',
               'Lymphocytes', 'WCC', 'CRP', 'Iron', 'Transf', '%Trans sat', 'Ferritin', 'CK', ]

# Iterate over each column_list and replace NaN values with the column's median
for column in column_list:
    if df[column].isnull().any():  # Check if there is any NaN value in the column
        median_value = df[column].median()
        df[column] = df[column].fillna(median_value)

# Now 'df' has the missing values in the selected columns replaced with their respective median values
```

```python
## Find out if the median pain ratings at timepoints T0 and T2 are statistically different

# Get unique timepoints
timepoints = df['Timepoint'].unique()

# Perform Mann-Whitney U test
test_results = []

# Iterate over each pair of timepoints
for i in range(len(timepoints)):
    for j in range(i + 1, len(timepoints)):
        # Filter the pain rating data for each timepoint and drop NaN values
        data1 = df[df['Timepoint'] == timepoints[i]]['Pain rating'].dropna()
        data2 = df[df['Timepoint'] == timepoints[j]]['Pain rating'].dropna()

        # Perform the Mann-Whitney U test
        statistic, p_value = mannwhitneyu(data1, data2, alternative='two-sided')

        # Append the test results to the list
        test_results.append((timepoints[i], timepoints[j], statistic, p_value))

# Print the Mann-Whitney U test results
for result in test_results:
    timepoint1, timepoint2, statistic, p_value = result
    print(f"Mann-Whitney U test between Timepoint {timepoint1} and Timepoint {timepoint2}: statistic={statistic:.2f}, p-value={p_value:.24f}")
    print("\n")
```

```python
## Find out if the pain rating distributions at timepoints T0 and T2 are statistically dissimilar

# Get unique timepoints
timepoints = df['Timepoint'].unique()

# Perform Kolmogorov-Smirnov test
test_results = []

# Iterate over each pair of timepoints
for i in range(len(timepoints)):
    for j in range(i + 1, len(timepoints)):
        # Filter the pain rating data for each timepoint
        data1 = df[df['Timepoint'] == timepoints[i]]['Pain rating']
        data2 = df[df['Timepoint'] == timepoints[j]]['Pain rating']

        # Perform the Kolmogorov-Smirnov test
        statistic, p_value = ks_2samp(data1, data2)

        # Append the test results to the list
        test_results.append((timepoints[i], timepoints[j], statistic, p_value))

# Print the Kolmogorov-Smirnov test results
for result in test_results:
    timepoint1, timepoint2, statistic, p_value = result
    print(f"K-S test between Timepoint {timepoint1} and Timepoint {timepoint2}: statistic={statistic:.3f}, p-value={p_value:.18f}")
    print("\n")
```

```python
## Find out if the median biomarker values at timepoints T0 and T2 are statistically different

# Get unique timepoints
timepoints = df['Timepoint'].unique()

# Perform Mann-Whitney U test
test_results = []

# Iterate over each column in cols
for col in cols:
    # Iterate over each pair of timepoints
    for i in range(len(timepoints)):
        for j in range(i + 1, len(timepoints)):
            # Filter the data for each timepoint and drop NaN values
            data1 = df[df['Timepoint'] == timepoints[i]][col].dropna()
            data2 = df[df['Timepoint'] == timepoints[j]][col].dropna()

            # Perform the Mann-Whitney U test
            statistic, p_value = mannwhitneyu(data1, data2, alternative='two-sided')

            # Append the test results to the list
            test_results.append({
                'Column': col,
                'Timepoint1': timepoints[i],
                'Timepoint2': timepoints[j],
                'Statistic': statistic,
                'P-Value': round(p_value, 3)  # Round p-value to 3 decimal places
            })

# Convert the test results into a DataFrame
results_df_mann = pd.DataFrame(test_results)

# Sort KS values in descending order
results_df_mann_sorted = results_df_mann.sort_values('P-Value', ascending=True)

print(results_df_mann_sorted)
```

```python
## Find out if the biomarker distributions at timepoints T0 and T2 are statistically dissimilar

# Get unique timepoints
timepoints = df['Timepoint'].unique()

# Initialize a list to store test results
test_results = []

# Iterate over each column
for col in cols:
    for i in range(len(timepoints)):
        for j in range(i + 1, len(timepoints)):
            # Filter the data for each timepoint
            data1 = df[df['Timepoint'] == timepoints[i]][col]
            data2 = df[df['Timepoint'] == timepoints[j]][col]

            # Perform the Kolmogorov-Smirnov test
            statistic, p_value = ks_2samp(data1, data2)

            # Append the test results to the list
            test_results.append({
                'Column': col,
                'Timepoint1': timepoints[i],
                'Timepoint2': timepoints[j],
                'KS Statistic': statistic,
                'P-Value': f'{p_value:.3f}'
            })

# Convert the test results into a DataFrame
results_df = pd.DataFrame(test_results)

# Sort KS values in descending order
results_df_sorted = results_df.sort_values('KS Statistic', ascending=False)

print(results_df_sorted)
```

## STEP 4.5 = Evaluate the relation between biomarkers and pain rating at both timepoints

```python
# Prepare DataFrame for results
results_df = pd.DataFrame()

# List of biomarkers
biomarkers = ['CD4%', 'Platelets', 'Neutrophils', 'Transf', '%Trans sat']

# Loop over each biomarker
for biomarker in biomarkers:
    # Group by 'Pain rating' and 'Timepoint' and calculate medians
    medians = df.groupby(['Pain rating', 'Timepoint'])[biomarker].median().unstack()
    
    # Perform the Mann-Whitney U test for each pain rating available in both T0 and T2
    results = []
    for pain_rating in medians.index:
        if 'T0' in medians.columns and 'T2' in medians.columns:
            if not pd.isna(medians.loc[pain_rating, 'T0']) and not pd.isna(medians.loc[pain_rating, 'T2']):
                u_stat, p_value = mannwhitneyu(
                    df[(df['Timepoint'] == 'T0') & (df['Pain rating'] == pain_rating)][biomarker],
                    df[(df['Timepoint'] == 'T2') & (df['Pain rating'] == pain_rating)][biomarker],
                    alternative='two-sided'
                )
                results.append([pain_rating, medians.loc[pain_rating, 'T0'], medians.loc[pain_rating, 'T2'], u_stat, p_value])

    # Create a DataFrame for the current biomarker with results
    biomarker_df = pd.DataFrame(results, columns=[
        'Pain Rating', f'{biomarker} Median T0', f'{biomarker} Median T2', 
        f'{biomarker} U Statistic', f'{biomarker} P Value'
    ])
    
    # Merge or concatenate results
    if results_df.empty:
        results_df = biomarker_df
    else:
        results_df = pd.merge(results_df, biomarker_df, on='Pain Rating', how='outer')
```

## STEP 4.6 = Evaluate the influence of relevant biomarkers on pain rating

```python
#Create a new dataframe called df_change
df_change = df_T0[['Number']] # Get specimen ID column from timepoint T0

# Calculate the change in pain rating and relevant biomarker values between timepoints T0 to T2
df_change['pain_change'] = df_T2['Pain rating'].values - df_T0['Pain rating'].values
df_change['transf_change'] = df_T2['Transf'].values - df_T0['Transf'].values
df_change['%transsat_change'] = df_T2['%Trans sat'].values - df_T0['%Trans sat'].values
df_change['platelets_change'] = df_T2['Platelets'].values - df_T0['Platelets'].values
df_change['neutrophils_change'] = df_T2['Neutrophils'].values - df_T0['Neutrophils'].values
df_change['cd4_change'] = df_T2['CD4%'].values - df_T0['CD4%'].values

# Defining the dependent variable and independent variables
Y = df_change['pain_change']
X = df_change[['transf_change', 'cd4_change', 'platelets_change', 'neutrophils_change', '%transsat_change']]

# Setting Number as index column
df_change.set_index('Number', inplace = True)

# Adding a constant to the model (intercept)
X_const = sm.add_constant(X)
```

```python
print("Shape of X without constant:", X.shape)
print("Shape of X with constant:", X_const.shape)
```

```python
# Fit a robust regression model without adding a constant term
robust_model = sm.RLM(Y, X).fit()

# Display the summary
print(robust_model.summary())
```

```python
# Fit a robust regression model after adding a constant term
robust_model_const = sm.RLM(Y, X_const).fit()

# Display the summary
print(robust_model_const.summary())
```
